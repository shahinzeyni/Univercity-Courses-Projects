```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fifa=pd.read_csv('D:\\fifa.csv')
fifa_var=pd.read_csv('D:\\fifa_+variable_information.csv')
fifa.head()
```

```python
fifa.drop(columns=['Photo','Flag','Club Logo','Loaned From'],inplace=True)
```

```python
fifa.info()
```

```python
fifa['Value']=fifa['Value'].replace('\u20AC','',regex=True)
fifa['Release Clause']=fifa['Release Clause'].replace('\u20AC','',regex=True)
fifa['Wage']=fifa['Wage'].replace('\u20AC','',regex=True)
```

```python
fifa.head()
```

```python
fifa['Release Clause'].str.replace('€','').head()
```

# Replacing the K and M and Euro symbol with space

```python
fifa['Value']=fifa.Value.replace({'K':'','M':''},regex=True)
fifa.Value.head()

fifa['Wage'] = fifa.Wage.replace({'K':'','M':''},regex=True)
fifa.Wage.head()
```

```python
fifa['Release Clause']=fifa['Release Clause'].replace({'K':'','M':''},regex=True)
```

```python
fifa.head()
```

# Converting the the string type into the float

```python
type(fifa['Value'][0])
```

```python
fifa['Value']=fifa['Value'].astype('float')
type(fifa.Value[0])
```

```python
type(fifa.Wage[0])
```

```python
fifa['Wage']=fifa['Wage'].astype('float')
type(fifa.Wage[0])
```

```python
fifa['Joined']=pd.to_datetime(fifa['Joined'])
fifa['Joined']=fifa['Joined'].dt.year
```

```python
fifa['Joined']=fifa['Joined'].fillna(2018)
fifa.Joined.isna().sum()
```

```python
fifa['Joined']= fifa['Joined'].astype('int64') 
type(fifa.Joined[0])
```

```python
fifa['Contract Valid Until']=pd.to_datetime(fifa['Contract Valid Until'])
```

# Filling the NAN values with the mean imputation. 

```python
fifa['Height']=fifa.Height.str.replace("'",'.')
fifa['Height'] = fifa.Height.astype('float')
fifa['Height'] = fifa.Height.fillna(fifa.Height.mean())
```

```python
fifa['Weight']=fifa.Weight.str.replace('lbs','')
fifa['Weight']=fifa.Weight.astype('float')
fifa['Weight'] = fifa.Weight.fillna(fifa.Weight.mean())
```

```python
fifa['Release Clause'] = fifa['Release Clause'].astype('float')
fifa['Release Clause'] = fifa['Release Clause'].fillna(fifa['Release Clause'].mean())
fifa['Release Clause'].mean()
```
