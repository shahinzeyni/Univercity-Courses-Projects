Task1

```python
import pandas as pd
import os

masir_poshe = 'training_setA/'

list_faileha = [file for file in os.listdir(masir_poshe) if file.endswith('.psv')]

tedad_kol_bimaran = 0
tedad_bimaran_motala_be_sepsis = 0

for name_file in list_faileha:
    masir_file = os.path.join(masir_poshe, name_file)
    
    dataframe = pd.read_csv(masir_file, sep='|')
    
    if 'SepsisLabel' in dataframe.columns:
        tedad_bimaran_motala_be_sepsis += dataframe['SepsisLabel'].sum()
        
        tedad_kol_bimaran += len(dataframe)

darsad_bimaran_motala_be_sepsis = (tedad_bimaran_motala_be_sepsis / tedad_kol_bimaran) * 100 if tedad_kol_bimaran > 0 else 0

print(f"tedad kol bimaran: {tedad_kol_bimaran}")
print(f"tedad bimaran motala be sepsis: {tedad_bimaran_motala_be_sepsis}")
print(f"darsad bimaran motala be sepsis: {darsad_bimaran_motala_be_sepsis:.2f}%")
```

Task2
Gender Part

```python
import pandas as pd
import os

masir_poshe = 'training_setA/'

list_faileha = [file for file in os.listdir(masir_poshe) if file.endswith('.psv')]

tozi_e_jensiyat = {'M': 0, 'F': 0}

for name_file in list_faileha:
    masir_file = os.path.join(masir_poshe, name_file)
    
    dataframe = pd.read_csv(masir_file, sep='|')
    
    dataframe['Gender'] = dataframe['Gender'].map({0: 'M', 1: 'F'})
    
    tozi_e_jensiyat['M'] += dataframe[dataframe['Gender'] == 'M'].shape[0]
    tozi_e_jensiyat['F'] += dataframe[dataframe['Gender'] == 'F'].shape[0]

total_count = tozi_e_jensiyat['M'] + tozi_e_jensiyat['F']

print("\n--- Tozi-e Jensiyat (Be Sorat Darsad) ---")
print(f"Tedad mardan: {tozi_e_jensiyat['M']} ({(tozi_e_jensiyat['M'] / total_count) * 100:.2f}%)")
print(f"Tedad zanan: {tozi_e_jensiyat['F']} ({(tozi_e_jensiyat['F'] / total_count) * 100:.2f}%)")
```

Task2
میانگین و میانه

```python
import pandas as pd
import os

masir_poshe = 'training_setA/'

list_faileha = [file for file in os.listdir(masir_poshe) if file.endswith('.psv')]

tedad_kol_bimaran = 0
tedad_bimaran_motala_be_sepsis = 0
sen_bimaran_motala_be_sepsis = []
sen_bimaran_gheir_motala_be_sepsis = []

for name_file in list_faileha:
    masir_file = os.path.join(masir_poshe, name_file)
    
    dataframe = pd.read_csv(masir_file, sep='|')
    
    if 'SepsisLabel' in dataframe.columns and 'Age' in dataframe.columns:
        
        tedad_bimaran_motala_be_sepsis += dataframe['SepsisLabel'].sum()
        
        tedad_kol_bimaran += len(dataframe)
        
        sen_bimaran_motala_be_sepsis.extend(dataframe[dataframe['SepsisLabel'] == 1]['Age'])
        sen_bimaran_gheir_motala_be_sepsis.extend(dataframe[dataframe['SepsisLabel'] == 0]['Age'])

darsad_bimaran_motala_be_sepsis = (tedad_bimaran_motala_be_sepsis / tedad_kol_bimaran) * 100 if tedad_kol_bimaran > 0 else 0

miyangin_sen_bimaran_motala_be_sepsis = pd.Series(sen_bimaran_motala_be_sepsis).mean()
miyane_sen_bimaran_motala_be_sepsis = pd.Series(sen_bimaran_motala_be_sepsis).median()

miyangin_sen_bimaran_gheir_motala_be_sepsis = pd.Series(sen_bimaran_gheir_motala_be_sepsis).mean()
miyane_sen_bimaran_gheir_motala_be_sepsis = pd.Series(sen_bimaran_gheir_motala_be_sepsis).median()

print(f"tedad kol bimaran: {tedad_kol_bimaran}")
print(f"tedad bimaran motala be sepsis: {tedad_bimaran_motala_be_sepsis}")
print(f"darsad bimaran motala be sepsis: {darsad_bimaran_motala_be_sepsis:.2f}%")

print("\n--- Miyangin va Miyane Sen ---")
print(f"miyangin sen bimaran motala be sepsis: {miyangin_sen_bimaran_motala_be_sepsis}")
print(f"miyane sen bimaran motala be sepsis: {miyane_sen_bimaran_motala_be_sepsis}")
print(f"\nmiyangin sen bimaran gheir motala be sepsis: {miyangin_sen_bimaran_gheir_motala_be_sepsis}")
print(f"miyane sen bimaran gheir motala be sepsis: {miyane_sen_bimaran_gheir_motala_be_sepsis}")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

masir_poshe = 'training_setA/'

list_faileha = [file for file in os.listdir(masir_poshe) if file.endswith('.psv')]

modat_zaman_bastari_bimaran = []

for name_file in list_faileha:
    masir_file = os.path.join(masir_poshe, name_file)
    df = pd.read_csv(masir_file, sep='|')
    
    modat_zaman_bastari_bimaran.append(len(df))

plt.figure(figsize=(10, 6))
plt.hist(modat_zaman_bastari_bimaran, bins=range(1, max(modat_zaman_bastari_bimaran) + 1), edgecolor='black', alpha=0.7, color='yellow')
plt.title('توزیع طول مدت داده‌ای بیماران')
plt.xlabel('مدت زمان داده‌ها')
plt.ylabel('تعداد بیماران')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

masir_porshe = 'training_setA/'

list_fayl = [f for f in os.listdir(masir_porshe) if f.endswith('.psv')]

saat_ebtela_sepsis = []

for name_fayl in list_fayl:
    masir_fayl = os.path.join(masir_porshe, name_fayl)
    
    df = pd.read_csv(masir_fayl, sep='|')
    
    if df['SepsisLabel'].sum() > 0:
        saat_ebtela = df.loc[df['SepsisLabel'] == 1, 'ICULOS'].iloc[0]
        saat_ebtela_sepsis.append(saat_ebtela)

plt.figure(figsize=(10, 6))
plt.hist(saat_ebtela_sepsis, bins=20, color='yellow', edgecolor='black')
plt.title('Histogram of First Hour of Sepsis Onset')
plt.xlabel('Hour of Sepsis Onset')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
